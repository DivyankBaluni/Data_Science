**Project Summary**
* The Global Terrorism Database (GTD) is one of the most comprehensive open-source
databases on terrorist events worldwide. Maintained by the National Consortium for the Study of
Terrorism and Responses to Terrorism (START) at the University of Maryland, this dataset
catalogues over 181,691 domestic and international terrorist incidents that occurred between
1970 and 2017 across 205 countries, spanning 135 distinct attributes per record.
* This project performs in-depth Exploratory Data Analysis on the GTD to uncover global patterns,
trends, and insights related to the frequency, geography, methodology, and perpetrators of
terrorist activity. Through statistical summaries and data visualizations, the analysis examines
how terrorism has evolved over nearly five decades — from the politically motivated violence of
the Cold War era in South America and Western Europe to the surge in religiously extremist
activity in the Middle East and South Asia during the 2010s.
* Key findings from the EDA reveal that the Middle East & North Africa and South Asia are the
most heavily affected regions, together accounting for more than 53% of all recorded incidents.
Bombing and explosion is the most prevalent attack type (49%), and the Taliban, ISIL, Shining
Path, and Al-Shabaab are among the most active non-state actors. A dramatic escalation in
incidents is observed after 2011, peaking at 16,903 in 2014 — coinciding with the rise of ISIL
and ongoing conflicts in Syria, Iraq, Afghanistan, and Nigeria.
* The analysis also highlights the human toll of terrorism: over 411,868 fatalities and 523,869
injuries recorded in the dataset, with 2014 alone accounting for more than 44,000 deaths.
Success rates, weapon types, target categories, and temporal distributions are further explored
to provide a holistic view of the global terrorism landscape.
